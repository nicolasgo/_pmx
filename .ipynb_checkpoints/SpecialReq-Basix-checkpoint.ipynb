```python
# cap is the cap.. In 100MB
def func(x,cap=2):
    y= np.int16(x/(100*1024*1024))
    if y>cap:
        y=cap
    return y

df['T'] =df['total'].apply(lambda x: func(x,10))
df['T200'] =df['total'].apply(lambda x: func(x))
```

```python
import hashlib

saved = {}
rsaved = {}  # a reverse dictionary can be used to lookup original msisdn from hashed ids


def xx(x):
    if not saved.has_key(x):
        try:
            m = hashlib.md5()
            m.update(x)
            mh = '%.12s'%(m.hexdigest()[10:30])
            saved[x]=mh
            rsaved[mh]=x


        except:
            print type(x),x

#        print x,mh

    return saved[x]


#print len(df['msisdn'].unique())
#print len(df['ip'].unique())

# just to test
print xx('524111000443')
```
